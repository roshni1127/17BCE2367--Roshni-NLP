```python
import gensim
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import brown
nltk.download('brown')
```

```python
from gensim.models import word2vec
```

```python
brown_vectors = word2vec.Word2Vec(brown.sents())
```

```python
print(brown_vectors)
```

```python
words = list(brown_vectors.wv.vocab)
print(words)
```

```python
print(brown_vectors['Country'])
```

```python
from sklearn.decomposition import PCA
from matplotlib import pyplot
feature = brown_vectors[brown_vectors.wv.vocab]
pca = PCA(n_components=2)
model = pca.fit_transform(feature)
```

```python
pyplot.scatter(model[:, 0], model[:, 1])
for i, word in enumerate(words):
	pyplot.annotate(word, xy=(model[i, 0], model[i, 1]))
pyplot.show()
```
